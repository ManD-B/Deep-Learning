```python
import pandas as pd
import numpy as np
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARIMA
from datetime import datetime , timedelta
from sklearn.preprocessing import MinMaxScaler
```

```python
name = ['Gold','Silver','Nas','Dow','S&P','EUR','USD']
```

```python
step  = 10
```

```python
res = []
for i in name:
    temp = pd.read_csv('aa.csv', header=0, index_col=0, squeeze=True, sep=',', encoding = 'euc-kr',
                        usecols = ['date',i])
    temp.index = pd.to_datetime(temp.index, format='%Y%m%d')
    print(i)
    if i == 'date':
        pass
    elif i == 'EUR':
        model = ARIMA(temp, order=(2,0,1))
    elif i == 'USD' :
        model = ARIMA(temp, order=(3,1,2))
    else:
        model = ARIMA(temp, order=(2,1,2)) 
    model_fit = model.fit(trend='nc',full_output=True, disp=1)
    res.append(model_fit.forecast(steps=step)[0].tolist())
    print(model_fit.summary())
        
result = dict(zip(name,res))
```

```python
df_scaled = df_scaled.dropna()
```

```python
import tensorflow
import matplotlib.pyplot as plt
%matplotlib inline
from tensorflow import keras
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Embedding, Dense, LSTM
from sklearn.model_selection import train_test_split
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

tensorflow.random.set_seed(2020)
```

```python
a = ['Gold','Silver','EUR','USD','Dow','S&P']
feature = df_scaled[a]
target = df_scaled['Nas']
```

```python
x_train, x_test, y_train, y_test = train_test_split(feature,target, test_size =0.3,
                                                       random_state=2020, shuffle =False)
```

```python
model=Sequential([
    Dense(128, input_shape=(6,), activation="relu"),
    Dense(64, activation="relu"),
    Dense(32, activation="relu"),
    Dense(1)
])
model.summary()
```

```python
model.compile(loss="mse",optimizer="adam",metrics=["accuracy"])
```

```python
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4)
mc = ModelCheckpoint('model/rnn.h5', monitor='val_accuracy', 
                     verbose=1, mode='max', save_best_only=True)
```

```python
model.fit(x_train, y_train, validation_split=0.2, epochs=1000, batch_size=300,
                    verbose=2,callbacks=[mc, es])
```

```python
pred = model.predict(x_test)
```

```python
plt.figure(figsize=(15, 9))
plt.plot(y_test.values, label='actual')
plt.plot(pred, label='prediction')
plt.legend()
plt.show()
```
